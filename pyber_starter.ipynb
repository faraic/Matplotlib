```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#loading and reading the csv data files
city_data_df =pd.read_csv("data/city_data.csv")
ride_data_df =pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
CityRide_df=ride_data_df.merge(city_data_df,how="inner",on="city")

#creating a dataset with one obervation per city
UniqueCityRide_df=CityRide_df.drop_duplicates(subset="city",keep="first")
```

```python
# Display the data table for preview
UniqueCityRide_df.head()
```

## Bubble Plot of Ride Sharing Data

```python

ByCity=CityRide_df.groupby("city")
AverageFarepc=list(ByCity.mean()["fare"])

TotalNumberRidespc=list(ByCity.count()["ride_id"])


new=UniqueCityRide_df.copy()

new=new.sort_values("city")

new["Average Fare"]=AverageFarepc
new["Total Number Rides"]=TotalNumberRidespc
pd.options.display.float_format = "{:,.2f}".format



# Obtain the x and y coordinates for each of the three city types
Rural = new.type == "Rural"
UniqueCityRideRural_df=new[Rural]


Urban = new.type == "Urban"
UniqueCityRideUrban_df=new[Urban]


Suburban = new.type == "Suburban"
UniqueCityRideSuburban_df=new[Suburban]
ax1=UniqueCityRideUrban_df.plot.scatter(x="Total Number Rides",y="Average Fare",
                                    s=UniqueCityRideUrban_df["driver_count"]*10, color="Coral",
                                   label="Urban",edgecolor = "black",alpha=0.65,linewidths=0.8)
ax2=UniqueCityRideSuburban_df.plot.scatter(x="Total Number Rides",y="Average Fare",
                                       s=UniqueCityRideSuburban_df["driver_count"]*10, color="Skyblue",
                                       label="Suburban",edgecolor = "black",alpha=0.65,linewidths=0.8,ax=ax1)
ax3=UniqueCityRideRural_df.plot.scatter(x="Total Number Rides",y="Average Fare",
                                    s=UniqueCityRideRural_df["driver_count"]*10, color="Gold",
                                    label="Rural",edgecolor = "black",alpha=0.65,linewidths=0.8,ax=ax1)


plt.title("Pyber Ride Sharing Data (2016)")
plt.text(42,35,"Note:\nCircle size correlates with driver count per city")
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare($)')
plt.xlim([1, 41])
plt.ylim([18, 45])
plt.grid()
plt.legend(loc=1,markerscale=0.6, title="City Types")
plt.show()
plt.savefig("bubbleplot.png",bbox_inches="tight")
```

## Total Fares by City Type

```python
# Calculate Type Percents
ByType=CityRide_df.groupby("type")
TotalFares=ByType.sum()["fare"]

labels=TotalFares.index
colors=["Gold","Skyblue","Coral"]

plt.pie(TotalFares, labels=labels, colors=colors,explode=[0,0,0.1],autopct = "%1.1f%%",radius=1.5, shadow=True, 
        center=(0,0),startangle=180,pctdistance=0.6)
plt.title("% of Total Fares by City Type",pad=60)
plt.show()
plt.savefig("TotalFaresPie.png",bbox_inches="tight")
```

## Total Rides by City Type

```python
# Calculate Type Percents
ByType=CityRide_df.groupby("type")
TotalRides=ByType.count()["ride_id"]

labels=TotalRides.index
colors=["Gold","Skyblue","Coral"]

plt.pie(TotalRides, labels=labels, colors=colors,explode=[0,0,0.1],autopct = "%1.1f%%",radius=1.5, shadow=True, 
        center=(0,0),startangle=180,pctdistance=0.6)
plt.title("% of Total Rides by City Type",pad=60)
plt.show()
plt.savefig("TotalRidesPie.png",bbox_inches="tight")
```

## Total Drivers by City Type

```python
# Calculate Type Percents
ByType=UniqueCityRide_df.groupby("type")
TotalDrivers=ByType.sum()["driver_count"]

labels=TotalDrivers.index
colors=["Gold","Skyblue","Coral"]

plt.pie(TotalDrivers, labels=labels, colors=colors,explode=[0,0,0.4],autopct = "%1.1f%%",radius=1.5, shadow=True, 
        center=(0,0),startangle=180,pctdistance=0.6)
plt.title("% of Total Drivers by City Type",pad=60)
plt.show()
plt.savefig("TotalDriversPie.png",bbox_inches="tight")
```
